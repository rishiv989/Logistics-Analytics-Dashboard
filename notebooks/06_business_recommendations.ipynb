```python
import pandas as pd

fact = pd.read_csv(
    "../data/processed/fact_orders.csv"
)

fact["order_purchase_timestamp"] = pd.to_datetime(
    fact["order_purchase_timestamp"]
)

print(fact.shape)
```

```python
fact["delivery_bucket"] = pd.cut(
    fact["delivery_days"],
    bins=[0,5,10,20,500],
    labels=[
        "Fast",
        "Normal",
        "Slow",
        "Very Slow"
    ]
)
```

```python
total_revenue = fact["order_value"].sum()

top3_revenue = (
    fact.groupby("customer_state")["order_value"]
    .sum()
    .sort_values(ascending=False)
    .head(3)
    .sum()
)

share = (top3_revenue / total_revenue) * 100

print(f"Top 3 states contribute {share:.2f}% of total revenue")
```

```python
import pandas as pd

fact = pd.read_csv("../data/processed/fact_orders.csv")

# Any feature columns needed in this notebook
```

```python
total_revenue = fact["order_value"].sum()

top3_revenue = (
    fact.groupby("customer_state")["order_value"]
    .sum()
    .sort_values(ascending=False)
    .head(3)
    .sum()
)

share = (top3_revenue / total_revenue) * 100

print(f"Top 3 states contribute {share:.2f}% of total revenue")
```

```python
fact["delivery_bucket"] = pd.cut(
    fact["delivery_days"],
    bins=[0, 5, 10, 20, 500],
    labels=[
        "Fast",
        "Normal",
        "Slow",
        "Very Slow"
    ]
)

fact["delivery_bucket"].value_counts()
```

```python
fast_reviews = fact[
    fact["delivery_bucket"] == "Fast"
]["review_score"].mean()

slow_reviews = fact[
    fact["delivery_bucket"] == "Very Slow"
]["review_score"].mean()

print("Fast Delivery Review:", round(fast_reviews, 2))
print("Very Slow Delivery Review:", round(slow_reviews, 2))
```

```python
risk_states = (
    fact.groupby("customer_state")
    .agg(
        late_rate=("is_late", "mean"),
        avg_delivery=("delivery_days", "mean")
    )
)

risk_states["late_rate"] *= 100

risk_states.sort_values(
    "late_rate",
    ascending=False
).head(10)
```
