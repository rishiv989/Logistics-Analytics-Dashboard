```python
SELECT
    customer_state,
    ROUND(SUM(order_value)::numeric,2) AS revenue
FROM fact_orders
GROUP BY customer_state
ORDER BY revenue DESC
LIMIT 10;
```

```python
SELECT
    customer_state,
    ROUND(SUM(order_value)::numeric,2) AS revenue
FROM fact_orders
GROUP BY customer_state
ORDER BY revenue DESC
LIMIT 10;
```

```python
SELECT
    customer_state,
    ROUND(AVG(review_score)::numeric,2) AS avg_review
FROM fact_orders
GROUP BY customer_state
ORDER BY avg_review DESC;
```

```python
SELECT
    DATE_TRUNC('month', order_purchase_timestamp) AS month,
    ROUND(SUM(order_value)::numeric,2) AS revenue
FROM fact_orders
GROUP BY month
ORDER BY month;
```
